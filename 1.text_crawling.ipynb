### 한국경제

```python
# 데이터 불러오기
hk = pd.read_csv('./data/news_hk.csv')
hk = hk.drop('Unnamed: 0', axis=1)
hk
```

```python
chromedriver_autoinstaller.install()
options = webdriver.ChromeOptions()

driver = webdriver.Chrome(options=options)

hk_data = {'article':[]}

for link in hk['link']:
    try:
        driver.get(link)
        time.sleep(2)
        article = driver.find_element(By.ID, 'articletxt').text
        time.sleep(1)
        pro_article = clean_article(article)
        time.sleep(1)
        hk_data['article'].append(pro_article) 
    except Exception as e:
        print('오류:', e)
        hk_data['article'].append('')
```

```python
df = pd.DataFrame(hk_data)
df

# df.to_csv('./article/article_hk.csv', index=Trues)
```

```python
df.to_csv('./article/article_hk.csv')
```

### 매일경제

```python
mk = pd.read_csv('./data/news_mk.csv')
mk = mk.drop('Unnamed: 0', axis=1)
mk
```

```python
chromedriver_autoinstaller.install()
options = webdriver.ChromeOptions()

driver = webdriver.Chrome(options=options)

mk_data = {'article':[]}

for link in mk['link']:
    try:
        driver.get(link)
        time.sleep(2)
        article = driver.find_element(By.CLASS_NAME, 'news_cnt_detail_wrap')
        phrases = article.find_elements(By.TAG_NAME, 'p')
        for ph in phrases:
            time.sleep(1)
            pro_phrases = clean_article(ph.text)
            mk_data['article'].append(pro_phrases) 
            
    except Exception as e:
        print('오류:', e)
        hk_data['article'].append('')
```

```python
df2 = pd.DataFrame(mk_data)
df2
```

```python
df2.to_csv('./article/article_mk.csv')
```

### 연합뉴스TV

```python
yh = pd.read_csv('./data/news_yh.csv')
yh = yh.drop('Unnamed: 0', axis=1)
yh
```

```python
chromedriver_autoinstaller.install()
options = webdriver.ChromeOptions()

driver = webdriver.Chrome(options=options)

yh_data = {'article':[]}

for y in yh['link']:
    try:
        driver.get(y)
        time.sleep(2)

        article = driver.find_element(By.CLASS_NAME, 'article-body-text')
        pro_article = clean_article(article.text)

        # pro_article 내용을 줄 단위로 분리
        lines = pro_article.splitlines()

        # 마지막 8줄을 제외한 나머지 줄만 선택
        # 만약 전체 줄 수가 8줄 이하이면 빈 리스트가 됩니다.
        lines_to_print = lines[:-8]

        # 선택된 줄들을 다시 하나의 문자열로 합쳐서 출력
        yh_data['article'].append(lines_to_print)

        # print('\n'.join(lines_to_print))
        
    except Exception as e:
        print('에러', e)
```
